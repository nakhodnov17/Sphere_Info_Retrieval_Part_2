```python
from collections import defaultdict
import operator
```

```python
nodes = set()
hublikeness_old = defaultdict(float)
autlikeness_old = defaultdict(float)
hublikeness_new = defaultdict(float)
autlikeness_new = defaultdict(float)

edges_out = defaultdict(list)
edges_in = defaultdict(list)

with open("./data/link_graph_series.txt", 'r') as file:
    for line in file:
        left_node, right_node = line.rstrip().split('\t')
        
        nodes.add(left_node)
        nodes.add(right_node)
        
        hublikeness_old[left_node] = 1.        
        hublikeness_old[right_node] = 1.

        autlikeness_old[left_node] = 1.        
        autlikeness_old[right_node] = 1.
        
#         edges_out[left_node].add(right_node)        
#         edges_in[right_node].add(left_node)
        edges_out[left_node].append(right_node)        
        edges_in[right_node].append(left_node)
```

```python
HITS_iters = 5
```

```python
from copy import deepcopy
```

```python
for _ in range(HITS_iters):
    for node in nodes:
        hublikeness_new[node] = sum(autlikeness_old[key] for key in edges_out[node])      
        autlikeness_new[node] = sum(hublikeness_old[key] for key in edges_in[node])
    hublikeness_old = deepcopy(hublikeness_new)
    autlikeness_old = deepcopy(autlikeness_new)
```

```python
hub_top = list(reversed(sorted(hublikeness_old.items(), key=operator.itemgetter(1))))
aut_top = list(reversed(sorted(autlikeness_old.items(), key=operator.itemgetter(1))))
```
